```python
import cv2
image = cv2.imread("apple.jpg")
```

```python
import matplotlib.pyplot as plt
rgbimage = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(rgbimage)
```

```python
greyimage = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(greyimage)
```

```python
invertedimage=255-greyimage
plt.imshow(invertedimage)
```

```python
blurredimage = cv2.GaussianBlur(invertedimage,(21,21),0)
plt.imshow(blurredimage)
```

```python
blurrinverted=255-blurredimage
pencilsketch = cv2.divide(greyimage,blurrinverted,scale = 256.0)
plt.imshow(pencilsketch)
```
